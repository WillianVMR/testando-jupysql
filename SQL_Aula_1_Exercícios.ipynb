Exercício 12:
Selecione os 10 primeiros clientes da tabela "Customers", ordenados por nome em ordem alfabética.

```python
%%sql

SELECT *
FROM Customers
ORDER BY CompanyName
LIMIT 10
```

Exercício 13: Selecione todos os pedidos da tabela "Orders" que foram enviados através do transportadora com o ID 3 e cujo frete seja menor que $10

```python
%%sql

SELECT *
FROM Orders
WHERE ShipVia = 3 AND Freight < 10
```

Exercício 14:
Selecione todos os fornecedores da tabela "Suppliers" que têm o nome começando com a letra 'M' ou terminando com a letra 'A'.

```python
%%sql

SELECT *
FROM Suppliers
WHERE CompanyName LIKE 'M%' OR CompanyName LIKE '%A'
```

Exercício 15:
